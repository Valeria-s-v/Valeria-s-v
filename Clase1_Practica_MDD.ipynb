```python
import pandas as pd

# URL del dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
print(url)
```

```python
# Cargar el dataset directamente desde la URL
df = pd.read_csv(url)
df
```

```python
#Mostar los tipos de datos de cada columna
df.dtypes
```

```python
#Mostrar la cantidad total de datos
df.size
#Motrar cantidad total de filas
df.shape[0]
```

```python
#Cambiar el tipo de Datos de alguna columna
df['Age'] = df['Age'].astype(int)
```

```python
#Mostrar las columnas categoricas
df.select_dtypes(include=['object']).columns
```

```python
#Mostrar la cantidad de Nulos
df.isnull().sum()
#Eliminamos los valores nulos de la columna Age
df.dropna(subset=['Age'], inplace=True)
#df.dropna(inplace=True)
```

```python
df.shape[0]
```

```python
#Realizar un grafico de histograma de la columna Age
import matplotlib.pyplot as plt
df['Age'].hist()
plt.show()
```

```python
df.describe()
```

```python
#Reemplazar los valores nulos por el valor de la media
df['Age'].fillna(df['Age'].mean(), inplace=True)
#Reemplazar los valores nulos por el valor de la mediana

#df['Age'].fillna(df['Age'].mode()[0], inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
#Reemplazar los valores nulos por el valor de la moda
```

```python
#Realizar un grafico de histograma de la columna Age
import matplotlib.pyplot as plt
df['Age'].hist()
plt.show()
```

```python
# Filtrar filas donde la columna 'Age' es mayor que 30
filtered_df = df[df['Age'] > 30]

# Seleccionar columnas 'Name' y 'Age'
selected_columns = filtered_df[['Name', 'Age']]
selected_columns
```

```python
#Realizar un grafico de histograma de la columna Age
import matplotlib.pyplot as plt
df['Age'].hist()
plt.show()
```

```python
# Calcular el promedio de la columna 'Age'
mean_age = selected_columns['Age'].mean()

print(mean_age)
```

#CÃ³digo con Polars

```python
import polars as pl

# URL del dataset Titanic
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

# Cargar el dataset directamente desde la URL
df = pl.read_csv(url)
df


```

```python
# Filtrar filas donde la columna 'Age' es mayor que 30
filtered_df = df.filter(pl.col('Age') > 30)

# Seleccionar columnas 'Name' y 'Age'
selected_columns = filtered_df.select(['Name', 'Age'])

# Calcular el promedio de la columna 'Age'
mean_age = selected_columns['Age'].mean()

print(mean_age)
```
